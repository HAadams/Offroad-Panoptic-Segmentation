## Imports

```python
import gdown
import os
import sys


# Used for multiprocessing scripts
NUM_PROC = 20

# Add parent directory to path to enable access to other scripts)
module_path = os.path.abspath(os.path.join(".."))
if module_path not in sys.path:
    sys.path.append(module_path)
```

# Download Datasets

```python
dataset_path = os.path.join(os.getcwd(), "..", "datasets")
rugd_path = os.path.join(dataset_path, "RUGD")
rellis_path = os.path.join(dataset_path, "Rellis")

combined_path = os.path.join(dataset_path, "Combined")
extract_path = os.path.join(combined_path, "extracted")
extracted_images_path = os.path.join(extract_path, "images")
extracted_labels_path = os.path.join(extract_path, "labels")

images_path = os.path.join(combined_path, "images")
labels_path = os.path.join(combined_path, "labels")

print(extracted_images_path)
print(extracted_labels_path)
```

## RUGD

### Unzip RUGD

```python
rugd_image_zip_path = os.path.join(rugd_path, "RUGD_frames-with-annotations.zip")
rugd_label_zip_path = os.path.join(rugd_path, "RUGD_annotations.zip")

print(rugd_image_zip_path)
print(rugd_label_zip_path)

gdown.extractall(rugd_image_zip_path, extracted_images_path)
gdown.extractall(rugd_label_zip_path, extracted_labels_path)
```

### Convert images from .png to .jpg

```python
rugd_images_extract_path = os.path.join(extracted_images_path, "RUGD_frames-with-annotations")
print(rugd_images_extract_path)
rugd_labels_extract_path = os.path.join(extracted_labels_path, "RUGD_annotations")
print(rugd_labels_extract_path)
```

```python
from generateSplitDatasets import change_file_extensions

change_file_extensions(rugd_images_extract_path, ".png", ".jpg", NUM_PROC)
```

### Convert RUGD label colors to match Rellis

```python
from convertRugdToRellisFormat import convert_labels_to_rellis

convert_labels_to_rellis(rugd_labels_extract_path, NUM_PROC)
```

## Rellis

```python
rellis_image_zip_path = os.path.join(rellis_path, "Rellis_3D_pylon_camera_node.zip")
rellis_label_zip_path = os.path.join(rellis_path, "Rellis_3D_pylon_camera_node_label_color.zip")

print(rellis_image_zip_path)
print(rellis_label_zip_path)

gdown.extractall(rellis_image_zip_path, extracted_images_path)
gdown.extractall(rellis_label_zip_path, extracted_labels_path)
```
